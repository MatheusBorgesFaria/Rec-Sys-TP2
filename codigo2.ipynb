```python
import pandas as pd

ratings = pd.read_json('data/ratings.jsonl', lines=True)
content = pd.read_json('data/content.jsonl', lines=True)
targets = pd.read_csv('data/targets.csv')
```

```python
content
```

```python
ratings['Rating'] = pd.to_numeric(ratings['Rating'])
content.fillna('', inplace=True)  # Fill missing values in content
```

```python
content
```

```python
content['features'] = content['Genre'] + ' ' + content['Director'] + ' ' + content['Plot']
```

```python
content
```

```python
# Set the index before merging
ratings.set_index('ItemId', inplace=True)
content.set_index('ItemId', inplace=True)

# Use join instead of merge
user_profiles = ratings.join(content, on='ItemId', how='inner').reset_index()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
content_matrix = vectorizer.fit_transform(content['features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(content_matrix)
```
